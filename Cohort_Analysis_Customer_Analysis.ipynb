### **Cohort Analysis**

### **Acquisition month for each customer**

```python
# Merge df_customer and df_ecom_sales to link customer information with sales data
df_merged = pd.merge(df_ecom_sales, df_customer, on='CustomerID', how='left')

# Convert 'OrderDate' to datetime objects
df_merged['OrderDate'] = pd.to_datetime(df_merged['OrderDate'])

# Group by CustomerID and find the minimum OrderDate to get the first purchase date for each customer
df_first_purchase = df_merged.groupby('CustomerID')['OrderDate'].min().reset_index()
df_first_purchase.rename(columns={'OrderDate': 'FirstPurchaseDate'}, inplace=True)

# Extract the 'first purchase month' from the minimum OrderDate
df_first_purchase['CohortMonth'] = df_first_purchase['FirstPurchaseDate'].dt.to_period('M')

# Filter df_first_purchase to include only CohortMonth from 2022
df_first_purchase = df_first_purchase[df_first_purchase['CohortMonth'].dt.year == 2022]

display(df_first_purchase.head())
```

```python
# Merge 'CohortMonth' into df_merged (re-run after filtering df_first_purchase)
df_merged = pd.merge(df_merged.drop(columns=['CohortMonth'], errors='ignore'),
                     df_first_purchase[['CustomerID', 'CohortMonth']], on='CustomerID', how='left')

# Extract 'OrderMonth' from 'OrderDate'
df_merged['OrderMonth'] = df_merged['OrderDate'].dt.to_period('M')

# Filter df_merged to only include orders placed in 2022
df_merged = df_merged[df_merged['OrderMonth'].dt.year == 2022]

display(df_merged.head())
```

```python
# Drop rows where CohortMonth is NaT (customers whose first purchase was not in 2022)
df_cohort = df_merged.dropna(subset=['CohortMonth'])

display(df_cohort.head())
```

```python
# Calculate MonthSinceFirstPurchase (0-indexed)
df_cohort.loc[:, 'MonthSinceFirstPurchase'] = (df_cohort['OrderMonth'].dt.year - df_cohort['CohortMonth'].dt.year) * 12 + \
                                       (df_cohort['OrderMonth'].dt.month - df_cohort['CohortMonth'].dt.month)

# Group by CohortMonth and MonthSinceFirstPurchase to count unique customers
cohort_counts = df_cohort.groupby(['CohortMonth', 'MonthSinceFirstPurchase'])['CustomerID'].nunique().reset_index()

# Create the pivot table for cohort analysis
cohort_pivot = cohort_counts.pivot_table(index='CohortMonth', columns='MonthSinceFirstPurchase', values='CustomerID')

display(cohort_pivot)
```

### **Retention rate of customer**

```python
# Calculate retention rate
cohort_sizes = cohort_pivot.iloc[:,0]
retention_matrix = cohort_pivot.divide(cohort_sizes, axis=0)

display(retention_matrix.round(3) * 100)
```

- Initial Drop-off: For most cohorts, there's a significant drop in retention after the first month (Month 0 to Month 1). This is very common, as some customers might make a one-time purchase. For example, the January 2022 cohort drops from 100% to 0.4% retention in the first month.

- Varying Retention by Cohort: Some cohorts show slightly better initial retention than others. For instance, the February 2022 cohort has a 1.9% retention rate in Month 1, which is higher than January's 0.4%.

- Longer-term Retention: As time progresses, the retention rates generally continue to decrease, with most cohorts showing very low retention percentages after a few months. For example, by Month 7, the January cohort retains only 2.4% of its original customers.

### **Compare the retention rate between the Corporate and Consumer groups**

##### **Cohort Analysis for Corporate Segment**

```python
# Check unique segments in df_cohort
segments = df_cohort['Segment'].unique()
print(f"Unique Segments: {segments}")
```

```python
# Filter df_cohort for Corporate segment
df_cohort_corporate = df_cohort[df_cohort['Segment'] == 'Corporate'].copy()

# Calculate MonthSinceFirstPurchase for Corporate segment
df_cohort_corporate.loc[:, 'MonthSinceFirstPurchase'] = (df_cohort_corporate['OrderMonth'].dt.year - df_cohort_corporate['CohortMonth'].dt.year) * 12 + \
                                                 (df_cohort_corporate['OrderMonth'].dt.month - df_cohort_corporate['CohortMonth'].dt.month)

# Group by CohortMonth and MonthSinceFirstPurchase to count unique customers for Corporate segment
cohort_counts_corporate = df_cohort_corporate.groupby(['CohortMonth', 'MonthSinceFirstPurchase'])['CustomerID'].nunique().reset_index()

# Create the pivot table for Corporate segment
cohort_pivot_corporate = cohort_counts_corporate.pivot_table(index='CohortMonth', columns='MonthSinceFirstPurchase', values='CustomerID')

# Calculate retention rate for Corporate segment
cohort_sizes_corporate = cohort_pivot_corporate.iloc[:,0]
retention_matrix_corporate = cohort_pivot_corporate.divide(cohort_sizes_corporate, axis=0)

print("Corporate Segment Retention Matrix:")
display(retention_matrix_corporate.round(3) * 100)
```

**Corporate Segment Retention Matrix:**

- Initial Drop-off: Similar to the overall cohort analysis, there's a significant drop after Month 0. For example, the January 2022 Corporate cohort drops from 100% to around 1.2% by Month 2, and some months even show no retention in Month 1.
- Varying Monthly Retention: Retention rates for the Corporate segment appear to fluctuate more month-to-month. Some months show slightly higher retention at later stages (e.g., Month 8 for Jan cohort is 3.6%).

##### **Cohort Analysis for Consumer Segment**

```python
# Filter df_cohort for Consumer segment
df_cohort_consumer = df_cohort[df_cohort['Segment'] == 'Consumer'].copy()

# Calculate MonthSinceFirstPurchase for Consumer segment
df_cohort_consumer.loc[:, 'MonthSinceFirstPurchase'] = (df_cohort_consumer['OrderMonth'].dt.year - df_cohort_consumer['CohortMonth'].dt.year) * 12 + \
                                               (df_cohort_consumer['OrderMonth'].dt.month - df_cohort_consumer['CohortMonth'].dt.month)

# Group by CohortMonth and MonthSinceFirstPurchase to count unique customers for Consumer segment
cohort_counts_consumer = df_cohort_consumer.groupby(['CohortMonth', 'MonthSinceFirstPurchase'])['CustomerID'].nunique().reset_index()

# Create the pivot table for Consumer segment
cohort_pivot_consumer = cohort_counts_consumer.pivot_table(index='CohortMonth', columns='MonthSinceFirstPurchase', values='CustomerID')

# Calculate retention rate for Consumer segment
cohort_sizes_consumer = cohort_pivot_consumer.iloc[:,0]
retention_matrix_consumer = cohort_pivot_consumer.divide(cohort_sizes_consumer, axis=0)

print("Consumer Segment Retention Matrix:")
display(retention_matrix_consumer.round(3) * 100)
```

**Consumer Segment Retention Matrix:**

- Initial Drop-off: The Consumer segment also experiences a sharp initial drop. The January 2022 Consumer cohort, for instance, has a NaN for Month 1, then shows 0.8% retention in Month 4, indicating a very low initial engagement post-acquisition.
- Relatively Consistent, but Low, Retention: After the initial drop, the retention rates tend to remain very low, often below 5% for most cohorts across the subsequent months.


**Comparison Insights:**

- Low Overall Retention for Both: Both Corporate and Consumer segments show generally low retention rates, suggesting a common challenge in customer engagement after the first purchase.
- Corporate vs. Consumer Early Retention: There doesn't appear to be a consistently clear winner in early retention (Month 1-3). Some Corporate cohorts show slightly better early retention, while others have NaN values, indicating zero retained customers in those early months. Consumer cohorts also show similar patterns.
- Data Sparsity: Many cells in both matrices show NaN values, especially in later months for various cohorts. This could mean either zero retained customers or simply that the data period for those specific cohort-month combinations has not yet occurred or is too small to show a value. This makes it harder to draw definitive long-term conclusions.

### **Cohort Analysis by Annual Income**

```python
# Display description of AnnualIncome in df_customer to determine bins
display(df_customer['AnnualIncome'].describe())
```

From the description, the annual income ranges from 10,000 to 170,000, with a mean of around 57,000. I'll create income groups using these quantiles:

```python
# Define income bins and labels based on quantiles and domain knowledge
income_bins = [0, 30000, 60000, 90000, 120000, df_customer['AnnualIncome'].max()]
income_labels = ['Low Income (0-30k)', 'Mid Income (30k-60k)', 'Upper-Mid Income (60k-90k)', 'High Income (90k-120k)', 'Very High Income (>120k)']

# Create 'IncomeGroup' column in df_customer
df_customer['IncomeGroup'] = pd.cut(df_customer['AnnualIncome'], bins=income_bins, labels=income_labels, right=False)

# Merge 'IncomeGroup' into df_cohort
df_cohort_income = pd.merge(df_cohort, df_customer[['CustomerID', 'IncomeGroup']], on='CustomerID', how='left')

# Display the first few rows with the new IncomeGroup
display(df_cohort_income.head())
```

```python
for group in income_labels:
    print(f"\n--- Cohort Analysis for Income Group: {group} ---")

    df_current_income_group = df_cohort_income[df_cohort_income['IncomeGroup'] == group].copy()

    if df_current_income_group.empty:
        print("No data for this income group.")
        continue

    # Calculate MonthSinceFirstPurchase (already exists in df_cohort_income, but re-calculate for safety)
    # df_current_income_group.loc[:, 'MonthSinceFirstPurchase'] = (df_current_income_group['OrderMonth'].dt.year - df_current_income_group['CohortMonth'].dt.year) * 12 + \
    #                                                        (df_current_income_group['OrderMonth'].dt.month - df_current_income_group['CohortMonth'].dt.month)

    # Group by CohortMonth and MonthSinceFirstPurchase to count unique customers
    cohort_counts_income = df_current_income_group.groupby(['CohortMonth', 'MonthSinceFirstPurchase'])['CustomerID'].nunique().reset_index()

    # Create the pivot table for the current income group
    cohort_pivot_income = cohort_counts_income.pivot_table(index='CohortMonth', columns='MonthSinceFirstPurchase', values='CustomerID')

    # Calculate retention rate for the current income group
    cohort_sizes_income = cohort_pivot_income.iloc[:,0]
    retention_matrix_income = cohort_pivot_income.divide(cohort_sizes_income, axis=0)

    display(retention_matrix_income.round(3) * 100)
```

**Overall Trend:**

- Initial Drop-off: Consistent with the overall and segment-specific analyses, all income groups experience a significant drop in retention immediately after the first purchase (Month 0 to Month 1). This indicates that many customers across all income levels make a single purchase.
- Low Long-Term Retention: Generally, long-term retention remains quite low across all income brackets. Very few customers continue to make purchases after the first few months.

**Comparison by Income Group:**

- Low Income (0-30k) & Mid Income (30k-60k): These groups show very similar patterns of low and often sporadic retention after the initial month. There isn't a clear advantage in sustainability between these two groups.
- Upper-Mid Income (60k-90k): While still facing an initial drop, this group sometimes exhibits slightly more consistent, albeit still low, retention in later months compared to the lower income groups. They might have a marginally better likelihood of repeat purchases over a longer period.
- High Income (90k-120k) & Very High Income (>120k): For these higher income brackets, the retention rates in later months (e.g., Month 4 onwards) appear to be slightly more stable or show higher percentages in some instances, though the absolute numbers of retained customers remain small. This suggests that while fewer in number, customers in these higher income groups might have a slightly higher propensity for longer-term engagement once they pass the initial drop-off phase. However, the data can be sparse for these groups in later months, making definitive conclusions challenging.

**Conclusion on Most Sustainable Group:**

- Based on the analysis, customers in the Upper-Mid Income (60k-90k), High Income (90k-120k), and Very High Income (>120k) groups tend to show a marginally more sustainable retention pattern compared to the lower income groups, particularly in the later months. While overall retention is low for everyone, the higher income groups appear to have a slightly better, albeit still fragile, long-term engagement. This indicates that efforts to foster loyalty might yield better returns in these segments, as they show a slightly higher propensity for continued interaction after the initial purchase.

**It's important to note that the retention numbers are generally very low for all groups, suggesting a broader challenge in customer retention that extends beyond just income levels.**
